### Hodgkin Huxley Simulation

To see the simulation code please checkout `Neuronal_Dynamics/LIF.py` which has the class `HodgkinHuxleyNeuron` which is used to simulate the Hodgkin Huxley model.


```python
from Neuronal_dynamics import *
```

```python
Neuron = HodgkinHuxley_Neuron(1, 120/100, 36/100, -12, 10.6, 0.3/100, 115)
```

```python
v,m, n, h, t = Neuron.simulate(1000, 0.02, 0.1)
plt.plot(t,v)
```

```python
plt.plot(t,m)
plt.plot(t,n)
plt.plot(t,h)
plt.legend([ "m", "n", "h"])
```

## Simulation for different Values of $g_k$

```python
Neuron1 = HodgkinHuxley_Neuron(1, 120/100, 0.02, -12, 10.6, 0.3/100, 115)
Neuron2 = HodgkinHuxley_Neuron(1, 120/100, 0.36, -12, 10.6, 0.3/100, 115)
Neuron3 = HodgkinHuxley_Neuron(1, 120/100, 0.72, -12, 10.6, 0.3/100, 115)
```

```python
v,m, n, h, t = Neuron1.simulate(1000, 0, 0.1)
plt.plot(t,v)
v, m, n, h, t = Neuron2.simulate(1000, 0, 0.1)
plt.plot(t,v)
v, m ,n, h, t = Neuron3.simulate(1000, 0, 0.1)
plt.plot(t,v)
plt.legend(["g_k = 0.02", "g_k = 0.36", "g_k = 0.72"])
plt.title("Voltage for different g_k values")
```
